# **Exercice 1.1 **

```python
# EXERCICE 1.1 : Question 1 et Question 2

# 1. Saisie du montant des achats
montant = float(input("Montant des achats (€) : "))

# 2. Test de la condition de livraison gratuite
if montant >= 1000:
    print("LIVRAISON GRATUITE")
    reduction = montant * 0.10        # 10 % de réduction
    total = montant - reduction
    print(f"Réduction accordée : {reduction:.2f}€")
else:
    print("LIVRAISON STANDARD (15€)")
    frais_livraison = 15.0
    total = montant + frais_livraison
    print(f"Frais de livraison : {frais_livraison:.2f}€")

# 3. Affichage du total à payer
print(f"Total à payer : {total:.2f}€")
```

# **Exercice 1.3**

```python
# EXERCICE : Comptage des clients fidèles

# 1. Liste des montants dépensés
depenses = [120, 30, 560, 999, 1020, 480, 700, 50]

# 2. Initialisation du compteur
nb_fideles = 0

print("DÉPENSES DES CLIENTS")
print("=" * 30)

# 3. Parcours de la liste
for montant in depenses:
    print(f"Montant du client : {montant}€")
    if montant >= 500:
        nb_fideles += 1

# 4. Calculs finaux
nb_total = len(depenses)
pourcentage = (nb_fideles / nb_total) * 100

# 5. Affichage du récapitulatif
print("=" * 30)
print(f"Nombre total de clients        : {nb_total}")
print(f"Nombre de clients fidèles      : {nb_fideles}")
print(f"Pourcentage de clients fidèles : {pourcentage:.2f}%")
```

# **Exercice 1.4**

```python
# EXERCICE 1.4 : Suivi des ventes jusqu'à l'objectif

# 1. Objectif de chiffre d'affaires
objectif = 3000.0

# 2. Initialisation
ventes_cumulees = 0.0
numero_vente = 1

print(f"Objectif de chiffre d'affaires : {objectif:.2f}€")
print("-" * 40)

# 3. Boucle while : tant que l'objectif n'est pas atteint
while ventes_cumulees < objectif:
    montant_vente = float(input(f"Montant de la vente n°{numero_vente} (€) : "))

    # Ajout au total
    ventes_cumulees += montant_vente
    numero_vente += 1

    # Affichage du suivi
    reste = objectif - ventes_cumulees
    print(f"Ventes cumulées : {ventes_cumulees:.2f}€")
    print(f"Reste à atteindre : {reste:.2f}€")
    print("-" * 30)

# 4. Bilan final
print("OBJECTIF ATTEINT OU DÉPASSÉ !")
print(f"Total des ventes : {ventes_cumulees:.2f}€")
print(f"Nombre de ventes : {numero_vente - 1}")
print("=" * 40)
```

# **Exercice 1.5**

```python
# EXERCICE 1.5 : Scoring de satisfaction client

# 1. Nombre de clients
n = int(input("Combien de clients voulez-vous saisir ? "))

# 2. Liste pour stocker les notes
notes = []

# Compteurs
nb_tres_satisfaits = 0
nb_satisfaits = 0
nb_insatisfaits = 0

# 3. Boucle de saisie
for i in range(1, n + 1):

    # Saisie avec contrôle
    while True:
        note = float(input(f"Note de satisfaction du client n°{i} (entre 0 et 10) : "))
        if 0 <= note <= 10:
            break
        else:
            print("Erreur : la note doit être comprise entre 0 et 10.")

    # Ajout dans la liste
    notes.append(note)

    # Classification
    if note >= 8:
        print("→ Client très satisfait")
        nb_tres_satisfaits += 1
    elif note >= 5:
        print("→ Client satisfait")
        nb_satisfaits += 1
    else:
        print("→ Client insatisfait")
        nb_insatisfaits += 1

    print("-" * 30)

# 4. Calculs finaux
moyenne = sum(notes) / len(notes)
meilleure_note = max(notes)
plus_faible_note = min(notes)

# 5. Affichage du bilan
print("\nBILAN GLOBAL")
print("=" * 30)
print(f"Nombre total de clients        : {n}")
print(f"Très satisfaits                : {nb_tres_satisfaits}")
print(f"Satisfaits                     : {nb_satisfaits}")
print(f"Insatisfaits                   : {nb_insatisfaits}")
print(f"Moyenne des notes              : {moyenne:.2f}/10")
print(f"Meilleure note                 : {meilleure_note}/10")
print(f"Plus faible note               : {plus_faible_note}/10")
print("=" * 30)
```

```python
# TP 1.1 - PARTIE A
# Téléchargement d'un fichier CSV depuis une URL

import requests

# 1. URL du fichier CSV à télécharger
url = "https://raw.githubusercontent.com/aiplanethub/Datasets/master/marketing_data.csv"

# 2. Envoyer une requête HTTP vers l'URL
r = requests.get(url)

# 3. Récupérer le contenu binaire du fichier
data = r.content

# 4. Sauvegarder le contenu dans un fichier local
nom_fichier = "adresses_clients.csv"

with open(nom_fichier, "wb") as f:
    f.write(data)

# 5. Message de confirmation
print(f"Fichier sauvegardé : {nom_fichier}")

```

```python
# Partie B question 3
# 3. Suppression des lignes où MntWines < 10
# On part du fichier sans "Education"
df_wines = df_sans_education.copy()

nb_total_avant = len(df_wines)

# Condition : garder uniquement les clients avec MntWines >= 10
df_wines_filtre = df_wines[df_wines["MntWines"] >= 10].copy()

nb_total_apres = len(df_wines_filtre)
nb_supprimes = nb_total_avant - nb_total_apres

print(f"Nombre de clients supprimés (MntWines < 10) : {nb_supprimes}")
print(f"Nombre de clients conservés : {nb_total_apres}")

# Sauvegarder le fichier filtré
fichier_wines = "clients_filtre_wines.csv"
df_wines_filtre.to_csv(fichier_wines, index=False)
print(f"Fichier sauvegardé : {fichier_wines}")
print("-" * 60)
```

```python
# Partie B question 4
# 4. Supprimer les clients dont MntFruits < 20 ET Recency <= 1
# On part du fichier déjà filtré sur MntWines
df_fruits_recency = df_wines_filtre.copy()

# Condition : lignes à supprimer
condition_suppr = (df_fruits_recency["MntFruits"] < 20) & (df_fruits_recency["Recency"] <= 1)

nb_lignes_suppr = condition_suppr.sum()

# On garde uniquement les lignes qui NE respectent PAS la condition
df_fr_filtre = df_fruits_recency[~condition_suppr].copy()

print(f"Nombre de lignes supprimées (MntFruits < 20 et Recency <= 1) : {nb_lignes_suppr}")

# Sauvegarder le nouveau fichier
fichier_fr_rec = "clients_filtre_fruits_recency.csv"
df_fr_filtre.to_csv(fichier_fr_rec, index=False)
print(f"Fichier sauvegardé : {fichier_fr_rec}")
```

# **Partie B – Histogrammes des montants d’achat**s

```python
# TP 1.2 - Partie B : Histogrammes des montants d’achats

import pandas as pd
import matplotlib.pyplot as plt

# 1. Charger le fichier Excel
fichier = "TP_1.2.xlsx"
df = pd.read_excel(fichier)

# 2. Création de deux histogrammes côte à côte
plt.figure(figsize=(12, 4))

# --- Histogramme des montants dépensés en vin ---
plt.subplot(1, 2, 1)   # 1 ligne, 2 colonnes, 1er graphique
plt.hist(df["MntWines"], bins=20)
plt.title("Répartition des montants dépensés en vin")
plt.xlabel("Montant dépensé en vin (MntWines)")
plt.ylabel("Nombre de clients")
plt.grid(axis="y", linestyle="--", alpha=0.5)

# --- Histogramme des montants dépensés en fruits ---
plt.subplot(1, 2, 2)   # 1 ligne, 2 colonnes, 2e graphique
plt.hist(df["MntFruits"], bins=20)
plt.title("Répartition des montants dépensés en fruits")
plt.xlabel("Montant dépensé en fruits (MntFruits)")
plt.ylabel("Nombre de clients")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

# **Commentaire :**
1. Répartition des montants dépensés en vin (MntWines)

La distribution est fortement asymétrique avec une concentration très élevée de clients dépensant moins de 200 DH.
Plus de 850 clients se situent dans la première classe (proche de 0 DH), ce qui indique que l’achat de vin est peu fréquent ou limité pour la majorité des clients.
À l’inverse, les montants élevés (jusqu’à 1500 DH) sont rares mais existent : cela révèle un segment de gros acheteurs, probablement des clients premium ou à revenu plus élevé.
La "queue" longue du graphique montre une forte dispersion des comportements :
certains clients achètent très peu, d’autres dépensent beaucoup → catégorie très hétérogène.
Le vin est un produit à fort potentiel de segmentation : un grand nombre de petits acheteurs et une minorité stratégique de gros consommateurs.

2. Répartition des montants dépensés en fruits (MntFruits).
La distribution est également asymétrique, mais les montants dépensés sont nettement plus faibles que pour le vin.
La majorité (près de 1200 clients) dépensent moins de 20 DH, ce qui traduit des achats petits et réguliers, probablement de consommation courante.
La dispersion est plus faible : les montants dépassent rarement 100 DH, et les valeurs extrêmes (jusqu’à 200 DH) sont très peu fréquentes.
Contrairement au vin, la "queue" du graphique est courte → comportement beaucoup plus homogène.
En résumé :
Les fruits sont une catégorie basique, peu différenciée, avec un comportement d’achat stable et peu variable entre les clients.

# **Partie C – Courbes temporelles**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier Excel
fichier = "TP_1.2.xlsx"
df = pd.read_excel(fichier)

# 1. Convertir Dt_Customer en datetime
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], errors="coerce")

# 2. Extraire l'année d'arrivée
df["Annee_arrivee"] = df["Dt_Customer"].dt.year

# 3. Calculer le nombre de nouveaux clients par année (trié)
clients_par_annee = (
    df["Annee_arrivee"]
      .value_counts()
      .sort_index()
)

# On récupère explicitement les années sous forme d'entiers
annees = clients_par_annee.index.astype(int)
nb_clients = clients_par_annee.values

print("Nombre de nouveaux clients par année :")
print(clients_par_annee)

# 4. Tracer la courbe temporelle
plt.figure(figsize=(10, 4))
plt.plot(annees, nb_clients, marker="o", linestyle="-")

plt.title("Évolution du nombre de nouveaux clients par année")
plt.xlabel("Année d'arrivée du client")
plt.ylabel("Nombre de nouveaux clients")
plt.grid(True)

# -> Ici on force les graduations de l’axe X à être exactement les années :
plt.xticks(annees)

plt.tight_layout()
plt.show()

```

# **Partie D – Boxplots**

```python
# TP 1.2 - Partie D : Boxplots des comportements d'achat

import pandas as pd
import matplotlib.pyplot as plt

# 1. Charger le fichier Excel
fichier = "TP_1.2.xlsx"      # adapter le chemin si besoin
df = pd.read_excel(fichier)

# 2. Créer un DataFrame ne contenant que les deux colonnes étudiées
colonnes_box = ["NumWebPurchases", "NumStorePurchases"]
df_box = df[colonnes_box].copy()

print("Aperçu de df_box :")
print(df_box.head())

# 3. Construire le boxplot des deux variables sur un même graphique
plt.figure(figsize=(8, 5))

df_box.boxplot()   # un boxplot pour chaque colonne

# Titre et labels
plt.title("Distribution des achats web et en magasin")
plt.ylabel("Nombre d'achats")          # axe des ordonnées (Y)
plt.xticks(rotation=15)                # rendre les noms de variables lisibles en X

plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

# **Commentaire**
Le graphique présente la distribution du nombre d’achats en ligne (NumWebPurchases) et du nombre d’achats en magasin (NumStorePurchases) pour l’ensemble des clients du dataset.

1. Comparaison des médianes

Les deux variables ont une médiane proche, autour de 4 à 5 achats.

Cela signifie que la moitié des clients effectue au moins 4 à 5 achats, que ce soit en magasin ou en ligne.

2. Dispersion (écart interquartile)

La boîte des achats en magasin (NumStorePurchases) est légèrement plus haute.
→ La variabilité des achats en magasin est un peu plus élevée que celle des achats en ligne.

Pour NumWebPurchases, la boîte est plus compacte :
→ les comportements sont plus homogènes.

3. Valeurs extrêmes

NumWebPurchases affiche plusieurs valeurs extrêmes élevées (jusqu’à 27–28 achats).
→ Quelques clients réalisent une activité web très intense, bien supérieure à la norme.

NumStorePurchases : très peu d’outliers, voire aucun.
→ Les achats en magasin sont plus réguliers et moins extrêmes.

4. Comportements généraux

La majorité des clients effectue entre 2 et 8 achats, quel que soit le canal.

Le web montre une queue longue vers le haut, révélant une minorité de clients très actifs en ligne.
